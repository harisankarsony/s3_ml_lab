```python
#pd2

import pandas as pd

s1 =pd.Series([10,20,30])
s2 =pd.Series([1,2,3])

print("1st series:")
print(s1)
print("\n2nd series:")
print(s2)
print("\nSum:")
print(s1+s2)
print("\nDifference:")
print(s1-s2)
print("\nProduct:")
print(s1*s2)
print("\nQuotient:")
print(s1/s2)
```

```python
#pd4

import pandas as pd
import numpy as np

dict1 ={"name":["hari","durga","eldhose","ajay","nibin","tony","savi","dibin","akhina","navneeth"],
       "score":[17,18,7,5,6,14,6,19,np.nan,np.nan],
       "attempts":[1,1,3,5,4,2,1,1,1,1],
       "qualify":["yes","yes","no","no","no","no","yes","yes","no","no"]}
dframe =pd.DataFrame(dict1)
print("Dataframe:\n",dframe)

#a)
print("\n")
print("Name and score:")
print(dframe[['name','score']])

#b)
print("\n")
print("Attempts greater than 2:")
print(dframe[dframe['attempts']>2])

#c)
print("\n")
print("No. of rows:",len(dframe.axes[0]))
print("No. of columns:",len(dframe.axes[1]))

#d)
print("\n")
print("Attempts less than 2 and score greater tha 10:")
print(dframe[(dframe['attempts']<2) & (dframe['score']>15)])

#e)
print("\n")
print("Rows with missing scores:")
print(dframe[dframe['score'].isnull()])
```

```python
#pd5

import pandas as pd
import numpy as np

dict2 ={"id":[1,2,3,4,5,6],
       "name":['hari','eldhose','ajay','tony','nibin','shabeeb'],
       "age":[20,26,30,34,np.nan,np.nan],
       "degree":["bca","bca","bsc","bsc",np.nan,np.nan],
       "cgpa":[7.0,8.0,6.5,9.5,np.nan,np.nan]}
dframe2 =pd.DataFrame(dict2)
print("Dataframe:\n",dframe2)

#a)

dframe2.insert(3,"class","mca")
print("\nAfter inserting class column:\n",dframe2)

#b)

print("\nNo. of columns:",len(dframe2.axes[1]))

#c)

print("\nTop 3 rows:")
print(dframe2.head(3))

print("\nLast 4 rows:")
print(dframe2.tail(4))

#e)

dframe2.pop('id')
print("\nAfter dropping id column:\n",dframe2)

#f)

print("\nname and ccgpa from records 2 to 5:")
dframe3 =dframe2.iloc[2:6]
print(dframe3[['name','cgpa']])

#g)

mean =dframe2['age'].mean()
dframe2['age'].fillna(value=mean,inplace=True)
print("\nAfter filling null age with mean:\n",dframe2)

#d)

dframe2 =dframe2.dropna()
print("\nAfter dropping rows with null values:\n",dframe2)
```

```python
#pd6

import pandas as pd

#a)
irisdf =df2 =pd.read_csv('datasets/iris.csv')
print("Iris dataframe:")
display(irisdf)

#b)
print("Descriptive statistics:")
display(irisdf.describe())

#c)
irisdf =irisdf.dropna()

#d)
classes =irisdf.species.unique()
for i in classes:
    print("Count of",i,"is",len(irisdf[(irisdf['species']==i)]))

#e)
irisdf['species']=irisdf['species'].map({'setosa':1,'versicolor':2,'virginica':3})
print("\nAfter mapping species class:")
print(irisdf['species'])

#f)
print("\nMean of petal length:",irisdf['petal_length'].mean())
print("Standard deviation of petal length:",irisdf['petal_length'].std())

#g)
print("\nDisplaying columns except species:")
display(irisdf.iloc[:,0:4])
```
