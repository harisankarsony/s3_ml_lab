```python
#pgm21 part1
#histogram and line plots
#Q(a)

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("datasets/iris.csv")
df = pd.DataFrame(df)

x = df["sepal_length"]
plt.hist(x,bins=20,color="green")

plt.title("sepal_length")
plt.xlabel("sepal_length")
plt.ylabel("count")
df.plot(kind="hist")
df.plot(kind="line")
df.plot(kind="line",subplots=True)
```

```python
#pgm21 part2

df.boxplot()
```

```python
#pgm21 p3

plt.rcParams["figure.figsize"] = (50,15)
df["sepal_length"].plot.bar()
```

```python
#pgm21 p4

plt.pie(df['species'].value_counts(),labels=['setosa','versicolor','virginica'],shadow=True)
```

```python
#pgm21 p4
#scatter plot
#Q(b)

df.plot(kind="scatter",x="sepal_length",y="petal_width")
```

```python
#pgm21 p4.1
#multiple scatter plot

ax1 = df.plot(kind='scatter', x='petal_length', y = 'petal_width', label='petal_length Vs petal_width', color='r')    
ax2 = df.plot(kind='scatter', x='petal_length', y = 'sepal_length', label='petal_length Vs sepal_length',color='g',ax=ax1)    
ax3 = df.plot(kind='scatter',  x='petal_length', y = 'sepal_width',label='petal_length Vs sepal_width', color='b',ax=ax2)
print(ax1 == ax2 == ax3)
```

```python
#pgm21 p5
#scatter matrix

from pandas.plotting import scatter_matrix
scatter_matrix(df,alpha=0.8,diagonal="kde")
```

```python
#pgm21 p6
#bubble plot

plt.scatter('petal_length','petal_width',s='sepal_length',alpha=0.5,data=df)
```

```python
#pgm21 p7

#Q(c)

sns.jointplot(x="sepal_length",y="sepal_width",data=df,height=8)
```

```python
#pgm21 p8

sns.boxplot(data=df)
```

```python
#p9

sns.pairplot(data=df,kind="scatter")
```

```python
sns.heatmap(df.corr(),annot=True)
```
